```python
import pandas as pd
```

```python
# Load the CSV files of the birds into a DataFrame
blue_teal_df = pd.read_csv('Resources/blue_winged_teal.csv')
flamingo_df = pd.read_csv('Resources/flamingo_and_others.csv')
hummingbirds_df = pd.read_csv('Resources/hummingbirds_and_others.csv')
owls_df = pd.read_csv('Resources/owls_and_others.csv')

# load the extra data to later be merged
bird_data_df = pd.read_csv('Resources/bird_data.csv')
age_df = pd.read_csv('Resources/age.csv')
sex_df = pd.read_csv('Resources/sex.csv')
```

```python
# Filter rows to include only the desired years
years_to_keep = list(range(1960, 1971)) + list(range(2010, 2023))

blue_filtered = blue_teal_df[blue_teal_df['EVENT_YEAR'].isin(years_to_keep)]
flamingo_filtered = flamingo_df[flamingo_df['EVENT_YEAR'].isin(years_to_keep)]
hummingbirds_filtered = hummingbirds_df[hummingbirds_df['EVENT_YEAR'].isin(years_to_keep)]
owls_filtered = owls_df[owls_df['EVENT_YEAR'].isin(years_to_keep)]
```

```python
blue_filtered.columns
```

```python
# Define a list of columns to delete
columns_to_delete = ['ISO_SUBDIVISION', 'EVENT_MONTH', 'EVENT_DAY', 'OTHER_BANDS', 'EXTRA_INFO', 'BAND_STATUS', 'HOW_OBTAINED', 'WHO_OBTAINED', 'REPORTING_METHOD', 'PRESENT_CONDITION', 'MIN_AGE_AT_ENC' ]

# Delete the specified columns
blue_filtered_df = blue_filtered.drop(columns=columns_to_delete)
flamingo_filtered_df = flamingo_filtered.drop(columns=columns_to_delete)
hummingbirds_filtered_df = hummingbirds_filtered.drop(columns=columns_to_delete)
owls_filtered_df = owls_filtered.drop(columns=columns_to_delete)

```

```python
# merge the extra data into the blue_teal dataset
blue_teal_merged = blue_filtered_df.merge(bird_data_df[['SPECIES_ID', 'SPECIES_NAME', 'SCIENTIFIC_NAME' ]], on='SPECIES_ID', how='left')
blue_teal_merged = blue_teal_merged.merge(age_df[['AGE_CODE', 'AGE_DESCRIPTION']], on='AGE_CODE', how='left')
blue_teal_merged = blue_teal_merged.merge(sex_df[['SEX_CODE', 'SEX_DESCRIPTION']], on='SEX_CODE', how='left')
```

```python
# merge the extra data into the flamingo dataset
flamingo_merged = flamingo_filtered_df.merge(bird_data_df[['SPECIES_ID', 'SPECIES_NAME', 'SCIENTIFIC_NAME' ]], on='SPECIES_ID', how='left')
flamingo_merged = flamingo_merged.merge(age_df[['AGE_CODE', 'AGE_DESCRIPTION']], on='AGE_CODE', how='left')
flamingo_merged = flamingo_merged.merge(sex_df[['SEX_CODE', 'SEX_DESCRIPTION']], on='SEX_CODE', how='left')
```

```python
flamingo_merged.columns
```

```python
# merge the extra data into the hummingbird dataset
hummingbirds_merged = hummingbirds_filtered_df.merge(bird_data_df[['SPECIES_ID', 'SPECIES_NAME', 'SCIENTIFIC_NAME' ]], on='SPECIES_ID', how='left')
hummingbirds_merged =hummingbirds_merged.merge(age_df[['AGE_CODE', 'AGE_DESCRIPTION']], on='AGE_CODE', how='left')
hummingbirds_merged =hummingbirds_merged.merge(sex_df[['SEX_CODE', 'SEX_DESCRIPTION']], on='SEX_CODE', how='left')
```

```python
hummingbirds_merged.columns
```

```python
owl_merged = owls_filtered_df.merge(bird_data_df[['SPECIES_ID', 'SPECIES_NAME', 'SCIENTIFIC_NAME' ]], on='SPECIES_ID', how='left')
owl_merged = owl_merged.merge(age_df[['AGE_CODE', 'AGE_DESCRIPTION']], on='AGE_CODE', how='left')
owl_merged = owl_merged.merge(sex_df[['SEX_CODE', 'SEX_DESCRIPTION']], on='SEX_CODE', how='left')
```
